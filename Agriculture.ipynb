### **Dividing Data into three dataframes**

```python
area_columns = [col for col in df.columns if 'AREA' in col]
production_columns = [col for col in df.columns if 'PRODUCTION' in col]
yield_columns = [col for col in df.columns if 'YIELD' in col]
drop_columns = ['FRUITS AREA (1000 ha)',
       'VEGETABLES AREA (1000 ha)', 'FRUITS AND VEGETABLES AREA (1000 ha)',
       'POTATOES AREA (1000 ha)', 'ONION AREA (1000 ha)',
       'FODDER AREA (1000 ha)']


selected_area_columns = [col for col in area_columns if col not in drop_columns]
df_area = df[['Year', 'State Name', 'Dist Name'] + selected_area_columns]
df_area['Total'] = df_area[selected_area_columns].sum(axis=1)

df_production = df[['Year', 'State Name', 'Dist Name'] + production_columns]
df_production['Total'] = df_production[production_columns].sum(axis=1)

df_yield = df[['Year', 'State Name', 'Dist Name'] + yield_columns]
df_yield['Total'] = df_yield[yield_columns].sum(axis=1)
```

### **Replacing negative values with 0**

```python

for col in df_area.columns:
    if df_area[col].dtype in ['int64', 'float64']:
        df_area[col] = df_area[col].replace(-1, 0)

for col in df_production.columns:
    if df_production[col].dtype in ['int64', 'float64']:
        df_production[col] = df_production[col].replace(-1, 0)

for col in df_yield.columns:
    if df_yield[col].dtype in ['int64', 'float64']:
        df_yield[col] = df_yield[col].replace(-1, 0)
```

### **Cleaning the column names**
*italicized text*

```python
import re

def clean_column_names(columns, keyword):
    new_columns = []
    for col in columns:
        cleaned_name = re.sub(rf" {keyword}.*|\(.*?\)", "", col).strip()
        new_columns.append(cleaned_name)
    return new_columns

df_area.columns = clean_column_names(df_area.columns, "AREA")
df_production.columns = clean_column_names(df_production.columns, "PRODUCTION")
df_yield.columns = clean_column_names(df_yield.columns, "YIELD")


print("Area Columns:", df_area.columns)
print("Production Columns:", df_production.columns)
print("Yield Columns:", df_yield.columns)
```

```python
# Compute correlation matrix
data = pd.read_csv("/content/ICRISAT-District Level Data.csv")
correlation_matrix = data[['RICE AREA (1000 ha)', 'RICE PRODUCTION (1000 tons)', 'RICE YIELD (Kg per ha)']].corr()
correlation_matrix
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)
sns.heatmap(correlation_matrix,annot=True)
```

```python
state_yield = data.groupby('State Name')[['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)']].mean().reset_index()

# Step 2: Melt the DataFrame to long format
state_yield_melted = state_yield.melt(id_vars='State Name',
                                        value_vars=['RICE YIELD (Kg per ha)', 'WHEAT YIELD (Kg per ha)'],
                                        var_name='Crop Type', value_name='Average Yield')

# Step 3: Create the bar plot using Seaborn
plt.figure(figsize=(12, 6))
sns.barplot(data=state_yield_melted, x='State Name', y='Average Yield', hue='Crop Type', palette='muted')
plt.title('Average Yield by State for Rice and Wheat')
plt.xlabel('State')
plt.ylabel('Average Yield (Kg per ha)')
plt.xticks(rotation=45)
plt.legend(title='Crop Type')
plt.tight_layout()
plt.show()
```

```python
df_crop_growth = df_production[['Year', 'RICE']].dropna()
df_crop_growth['Growth Rate (%)'] = df_crop_growth['RICE'].pct_change() * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Growth Rate (%)', data=df_crop_growth, marker='o')
plt.title('Yearly Growth Rate of Rice Production')
plt.xlabel('Year')
plt.ylabel('Growth Rate (%)')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Preparing the data for linear regression (for simplicity, using Rice Production)
df_rice = df_production[['Year', 'RICE']].dropna()

X = df_rice[['Year']].values  # Features (Years)
y = df_rice['RICE'].values  # Target (Production)

# Fit the model
model = LinearRegression()
model.fit(X, y)

# Predict future values (e.g., for the next 5 years)
future_years = np.array([[year] for year in range(2025, 2030)])
predictions = model.predict(future_years)

# Plotting
plt.figure(figsize=(12, 6))
sns.lineplot(x=df_rice['Year'], y=df_rice['RICE'], label='Historical Data')
sns.lineplot(x=future_years.flatten(), y=predictions, label='Predicted Production', linestyle='--')
plt.title('Rice Production Forecast')
plt.xlabel('Year')
plt.ylabel('Rice Production (1000 tons)')
plt.tight_layout()
plt.show()
```

```python
# Calculate the dependency ratio by dividing each crop production by the total production for each state
crop_dependency = dfp.set_index(["Year", "State Name", "Dist Name"]).drop(columns=["Total"])
dependency_ratio = crop_dependency.div(crop_dependency.sum(axis=1), axis=0).groupby("State Name").mean()
top_single_crop_states = dependency_ratio.idxmax(axis=1).value_counts().nlargest(5)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_single_crop_states.index, y=top_single_crop_states.values)
plt.title("Top States Most Dependent on a Single Crop")
plt.xlabel("Crop")
plt.ylabel("Number of Times a Single Crop is Dominant")
plt.show()
```

```python
# Compare yield variability for high-value crops and staple crops
staple_crops = ["RICE", "WHEAT"]
high_value_crops = ["SOYABEAN", "RAPESEED AND MUSTARD"]
staple_yield_variability = dfy.groupby("State Name")[staple_crops].std().mean(axis=1)
high_value_yield_variability = dfy.groupby("State Name")[high_value_crops].std().mean(axis=1)

plt.figure(figsize=(12, 6))
sns.kdeplot(staple_yield_variability, label="Staple Crops", fill=True, color="blue")
sns.kdeplot(high_value_yield_variability, label="High-Value Crops", fill=True, color="green")
plt.title("Yield Variability: High-Value vs Staple Crops Across States")
plt.xlabel("Yield Variability (Std Dev)")
plt.legend()
plt.show()
```

```python
key_crops = ["RICE", "WHEAT", "MAIZE"]
avg_yield = dfy.groupby(["Year", "State Name"])[key_crops].mean().reset_index()

# Get top 5 states by the highest average yield across the key crops
top_states = avg_yield.groupby("State Name")[key_crops].mean().sum(axis=1).nlargest(2).index
top_avg_yield = avg_yield[avg_yield["State Name"].isin(top_states)]

# Plot
plt.figure(figsize=(14, 8))
for crop in key_crops:
    sns.lineplot(data=top_avg_yield, x="Year", y=crop, hue="State Name", marker="o", legend=False)

plt.title("Yield Improvement Trends for Key Crops in Top 2 States")
plt.xlabel("Year")
plt.ylabel("Average Yield")

# Manually create the legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), title="State Name")

plt.show()
```

```python
import plotly.graph_objects as go

df = pd.read_csv('/content/ICRISAT-District Level Data.csv')

kharif_production_cols = [col for col in df.columns if 'KHARIF' in col and 'PRODUCTION' in col]
kharif_yield_cols = [col for col in df.columns if 'KHARIF' in col and 'YIELD' in col]
rabi_production_cols = [col for col in df.columns if 'RABI' in col and 'PRODUCTION' in col]
rabi_yield_cols = [col for col in df.columns if 'RABI' in col and 'YIELD' in col]

fig = go.Figure()

for col in kharif_production_cols:
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=df[col],
        mode='markers',
        name=f'Kharif Production - {col.split()[0]}',
        visible=True
    ))

for col in rabi_production_cols:
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=df[col],
        mode='markers',
        name=f'Rabi Production - {col.split()[0]}',
        visible=True
    ))

for col in kharif_yield_cols:
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=df[col],
        mode='markers',
        name=f'Kharif Yield - {col.split()[0]}',
        visible=False
    ))

for col in rabi_yield_cols:
    fig.add_trace(go.Scatter(
        x=df['Year'],
        y=df[col],
        mode='markers',
        name=f'Rabi Yield - {col.split()[0]}',
        visible=False
    ))

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="Production",
                     method="update",
                     args=[{"visible": [True] * (len(kharif_production_cols) + len(rabi_production_cols)) +
                                      [False] * (len(kharif_yield_cols) + len(rabi_yield_cols))},
                           {"title": "Crop Production Over Years"}]),
                dict(label="Yield",
                     method="update",
                     args=[{"visible": [False] * (len(kharif_production_cols) + len(rabi_production_cols)) +
                                      [True] * (len(kharif_yield_cols) + len(rabi_yield_cols))},
                           {"title": "Crop Yield Over Years"}])
            ]),
            direction="down",
            showactive=True,
        )
    ]
)

fig.update_layout(
    title="Crop Production and Yield (Rabi & Kharif Seasons)",
    xaxis_title="Year",
    yaxis_title="Value",
    legend_title="Crop Type and Season"
)

fig.show()
```

```python
crops = ["RICE", "WHEAT"]

df_all = pd.merge(dfa[['Year', 'State Name', 'Dist Name'] + crops],
                  dfp[['Year', 'State Name', 'Dist Name'] + crops],
                  on=['Year', 'State Name', 'Dist Name'], suffixes=('_Area', '_Production'))
df_all = pd.merge(df_all, dfy[['Year', 'State Name', 'Dist Name'] + crops],
                  on=['Year', 'State Name', 'Dist Name'], suffixes=('_Area', '_Yield'))

for crop in crops:
    df_all[f'{crop}_Prod_Growth'] = df_all[f'{crop}'].pct_change() * 100
    df_all[f'{crop}_Yield_Growth'] = df_all[f'{crop}'].pct_change() * 100

plt.figure(figsize=(14, 8))
for crop in crops:
    sns.lineplot(data=df_all, x='Year', y=f'{crop}_Prod_Growth', label=f'{crop} Production Growth', marker='o')
    sns.lineplot(data=df_all, x='Year', y=f'{crop}_Yield_Growth', label=f'{crop} Yield Growth', marker='x', linestyle='--')
plt.title("Annual Growth Rate of Crop Production and Yield Across States")
plt.xlabel("Year")
plt.ylabel("Growth Rate (%)")
plt.legend(title="Crops")
plt.show()
```
